### Import necessary libraries

```python
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

### Pull in the dataset

```python
# Load SciTLDR dataset
dataset = load_dataset("allenai/scitldr", "AIC")

print(dataset)
print(dataset["train"][0])
```

### Use a Pretrained model: Flan-5

```python
model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
def generate_summary(text, style="conversational"):
    prompt = f"Rewrite this scientific summary in {style} style: {text}"
    
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512)
    outputs = model.generate(
        **inputs,
        max_length=80,
        num_beams=4,
        early_stopping=True
    )
    
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
sample_text = dataset["train"][0]["source"][0]

print("Conversational:")
print(generate_summary(sample_text, "simple conversational English"))

print("\nGen-Z:")
print(generate_summary(sample_text, "funny Gen-Z language"))

print("\nCaveman:")
print(generate_summary(sample_text, "very simple caveman language"))
```
