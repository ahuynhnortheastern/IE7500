# Loading Datasets

### Import necessary libraries

```python
import json
import glob
import re
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Pull in the datasets

```python
# Testing to see how large the data files are

import os

# Get file size in bytes
# file_bytes = os.path.getsize("Data/dev.jsonl")
# file_bytes = os.path.getsize("Data/train.jsonl")
file_bytes = os.path.getsize("Data/test.jsonl")

# Convert to Megabytes (MB) for easier reading
file_mb = file_bytes / (1024 * 1024)

print(f"File size: {file_mb:.2f} MB")
```

```python
# Dictionary to hold the individual DataFrames
dfs_dict = {}

# Find and loop through every file
for filepath in glob.glob("Data/*.jsonl"):
    # Extract just the filename (e.g., "data_1") to use as a clean dictionary key
    filename = os.path.splitext(os.path.basename(filepath))[0]

    # Read the file and store it
    dfs_dict[filename] = pd.read_json(filepath, lines=True)

# Check to see all the files were successfully pulled
print(dfs_dict.keys())
```

```python
# Assess how many values are in each data frame
for name, values in dfs_dict.items():
    print("Dataset Name: ", name)
    print("Number of Rows in Data Frame: ", len(values))
```

```python
# Assess each df column
print(dfs_dict['dev'].columns.tolist())
print(dfs_dict['test'].columns.tolist())
print(dfs_dict['train'].columns.tolist())
```

```python
train_df = dfs_dict['train']
test_df = dfs_dict['test']
dev_df = dfs_dict['dev']

print("Missing values in train data:")
print(train_df.isnull().sum())

print("Missing values in dev data:")
print(dev_df.isnull().sum())

print("Missing values in test data:")
print(test_df.isnull().sum())
```
