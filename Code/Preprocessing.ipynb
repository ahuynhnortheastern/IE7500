# Loading Datasets

### Import necessary libraries

```python
import json
import glob
import re
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

### Pull in the datasets

```python
# Create an empty dictionary to store all the data frames
dfs_dict = {}

# Create a for loop to pull each file

# Pull all files that end with a .jsonl file type
for filepath in glob.glob("Data/*.jsonl"):

    # Extract the filename without the document type
    filename = filepath[:-6]
    filename = filename[5:]

    # Read the file
    dfs_dict[filename] = pd.read_json(filepath, lines=True)

# Check to see all the files were successfully pulled
print(dfs_dict.keys())
```

```python
# Assign a variable for each of the datasets
dev_df = dfs_dict['dev']
train_df = dfs_dict['train']
test_df = dfs_dict['test']

# Print the column names for each
print("Development Data Frame Columns: ", dev_df.columns.tolist())
print("Training Data Frame Columns: ", train_df.columns.tolist())
print("Testing Data Frame: ", test_df.columns.tolist())

# All of the columns are the same across all datasets
```

```python
# Check to see if there are any null values
print("Total # of Null Values in Development Data Frame: ", dev_df.isnull().sum())
print("Total # of Null Values in the Training Data Frame: ", train_df.isnull().sum())
print("Total # of Null Values in Testing Data Frame: ", test_df.isnull().sum())

# There are no null values in any of these datasets
```

```python
# Assess how many values are in each data frame
for name, values in dfs_dict.items():
    print("Dataset Name: ", name)
    print("Number of Rows in Data Frame: ", len(values))
```

```python
train_df = dfs_dict['train']
test_df = dfs_dict['test']
dev_df = dfs_dict['dev']

print("Missing values in train data:")
print(train_df.isnull().sum())

print("Missing values in dev data:")
print(dev_df.isnull().sum())

print("Missing values in test data:")
print(test_df.isnull().sum())
```

```python
# See what the dataset roughly looks like
# print(train_df.head(1))

print(train_df["source"][0])
print(train_df["target"][0])

# Save a line divider for print formatting
line = "--" * 40
print(line) 

# See what the source looks like
print("How many sentences are in the first source value: ", len(train_df['source'].iloc[0]))
print("How many sentences are in the first target value: ", len(train_df['target'].iloc[0]))

print(line)

# The source seems to be comprised of several sentences
# For the first index, there are 54 sentences within one list
# Since we are planning to leverage a RAG-based model, we need to combine the sentences together

# Create a function that joins all the sentences in the list
def join_sentences(sentences):
    return " ".join(sentences)

# For both the source and target column, join all the sentences
dev_df["source_paragraph"] = dev_df["source"].apply(join_sentences)
train_df["source_paragraph"] = train_df["source"].apply(join_sentences)
test_df["source_paragraph"] = test_df["source"].apply(join_sentences)

dev_df["target_paragraph"] = dev_df["target"].apply(join_sentences)
train_df["target_paragraph"] = train_df["target"].apply(join_sentences)
test_df["target_paragraph"] = test_df["target"].apply(join_sentences)

# Double check that it was joined properly
print(train_df["source_paragraph"][0])
print(train_df["target_paragraph"][0])
```
